```python
import ast as ast  #literal_eval convert string list to normal list
def convert(obj):
    Lst = []
    for i in ast.literal_eval(obj):
        Lst.append(i['name'])
    return Lst
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
 #literal_eval convert string list to normal list
def convert_Cast(obj):
    Lst = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            Lst.append(i['name'])
            count+=1;
        else:
            break; 
    return Lst
```

```python
movies['cast'] = movies['cast'].apply(convert_Cast)
```

```python
movies.head(1)
```

```python
def Find_Director(obj):
    Lst = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            Lst.append(i['name'])
            break
    return Lst
```

```python
movies['crew'] = movies['crew'].apply(Find_Director)
```

```python
movies.head(4)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
# now we remove space between all entities   our Aim : Sam Worthington = SamWorthington
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# we complete our preprocessing now what next ? 
# we do vectorozation we convert all our movies to vectors and then we recomendent closest top 5 vectorts to that the given vector movie 
# for this process we choose Bag of Words technique
```

```python
# Bag of Words Technique
# first appent all tag strings to a single string 
# take top 5000 words based on frequency on that string
# now check that 5000 words are there how many times in our tag1,tag2,tag3.....last movie Tag
# this table is a vector table.but we can ignore stop words like are,is,and...
```

```python
import re
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(txt):
    y = []
    txt = re.sub('[^a-zA-Z]', ' ', txt)  # remove punctuation & numbers
    for i in txt.split():
        y.append(ps.stem(i.lower()))
    return " ".join(y)
```

```python
ps.stem('loved') 
# this stem helps to reduce similar words like loved,love,loving this converts all this to love
```

```python
new_df['tags'][0]
```

```python
stem(new_df['tags'][0])
#see difference 
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
features = cv.get_feature_names_out()        # all top 5000 names
#print(list(features))
```

```python
vectors.shape
```

```python
# now we have movie vectors so we find distance between all vectors we use cosine distance
# distance inverse proportional to similirity
# dont use eculidian distance it will fail in higher dimensions 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
